```python
# 1. Import the necessary libraries and load the dataset.
import pandas as pd
import os

cwd = os.getcwd()
cwd
url='C:\\Users\\user\\Africa-Data-School-Curriculum\\notebooks/data/vgsales.csv'
df = pd.read_csv(url)

# 2. Display the first ten rows of the dataset.

df.head(10)

# 3. Display the shape and basic information of the dataset.

print(df.shape)
print(df.info())

# 4. Check for the missing data and handle it approprietly.

print(df.isna().sum())
df=df.dropna()

# 5. Display the top 5 video games by globalscale.

top5_global=df.nlargest(5, 'Global_Sales')
print(top5_global)

# 6. The total sales in North America for the action genre.

action_sales_na=df[df['Genre']=='Action']['NA_Sales'].sum()
print(action_sales_na)

# 7. The top 3 publishers by global sales.

top3_publisher= df.groupby('Publisher')['Global_Sales'].sum().nlargest(3)
print(top3_publisher)

# 8. The average sales in Europe in the 'Shooter' genre.

shooter_average_sales_eu=df[df['Genre']=='Shooter']['EU_Sales'].mean()
print(shooter_average_sales_eu)


# 9. The game with the highest sales in Japan in the 'Sports' genre.

highest_jp_sports=df[df['Genre']=='Sports'].nlargest(1, 'JP_Sales')
print(highest_jp_sports)

                                                     
                                                     








```
